## **Q5: Apply time series decomposition to your dataset to visualize the trend and seasonality**

```python
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

clmn = ["Agriculture", "Business services", "Construction", "Durable goods manufacturing", "Education and Health", "Finance", "Government", "Information", "Leisure and hospitality", "Manufacturing", "Mining and Extraction", "Nondurable goods manufacturing", "Other", "Self-employed", "Transportation and Utilities", "Wholesale and Retail Trade"]
i = 0
for (columnName, columnData) in df[clmn].iteritems(): 
    fig = plt.figure()
    ad = fig.add_subplot(111)
    ad.set_title("Trend and Seasonality of " + df[clmn].columns[i] + " Industry", fontsize = 15)
    graph = seasonal_decompose(df[columnName], model='additive')
    graph.seasonal.plot()
    graph.trend.plot()
    plt.legend(loc="upper right")
    plt.xlabel("Date in Years", fontsize = 15)
    i += 1
```

## **Q6: Visualize the seasonality of Agriculture, Health and Finance sector**

```python
new = ["Agriculture", "Education and Health", "Finance"]
i = 0
for (columnName, columnData) in df[new].iteritems():
    fig = plt.figure()
    ad = fig.add_subplot(111)
    ad.set_title("Seasonality of " + df[new].columns[i] + " Industry", fontsize = 25)
    graph = seasonal_decompose(df[columnName], model='additive')
    graph.seasonal.plot()
    plt.legend(loc="upper right")
    plt.xlabel("Date in Years", fontsize = 15)
    i += 1
```
